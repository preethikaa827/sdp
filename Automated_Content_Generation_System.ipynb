```python
# Cell 1: Install necessary libraries
!pip install transformers torch

# Cell 2: Import necessary libraries and load the model
from transformers import GPTJForCausalLM, GPT2Tokenizer
import torch

model_name = "gpt2"  # or use 'distilgpt2' for a smaller version
model = GPTJForCausalLM.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Cell 3 (Moved from Cell 4): Define the 'generate_content' function
def generate_content(prompt, max_length=100):
    """Generates text content using the pre-trained model.

    Args:
        prompt: The input prompt text.
        max_length: The maximum length of the generated text.

    Returns:
        The generated text.
    """
    # Encode the input prompt to tensor
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    # Generate text from the model
    with torch.no_grad():
        outputs = model.generate(inputs, max_length=max_length, num_return_sequences=1, no_repeat_ngram_size=2, temperature=0.7)

    # Decode the generated tokens into text
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return generated_text

```

```python
import re

def generate_content(prompt, max_length=100):
    # Encode the input prompt to tensor
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    # Generate text from the model
    with torch.no_grad():
        outputs = model.generate(inputs, max_length=max_length, num_return_sequences=1, no_repeat_ngram_size=2, temperature=0.7)

    # Decode the generated tokens into text
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return generated_text

def save_content_with_dynamic_filename(content):
    # Use the first 10 words from the generated content for the filename
    filename_part = " ".join(content.split()[:10])  # Extract the first 10 words
    filename_part = re.sub(r'\W+', '_', filename_part)  # Replace non-alphanumeric characters with underscores
    filename = f"generated_content_{filename_part}.txt"  # Generate the filename

    # Save the content to the file
    with open(filename, "w") as file:
        file.write(content)

    print(f"Content saved to {filename}")

# Example usage
prompt = "Write a short story about a young girl who discovers a hidden treasure in her backyard."
generated_content = generate_content(prompt, max_length=150)
save_content_with_dynamic_filename(generated_content)
```
